```python
from scipy.constants import pi, c, h, k
from math import e
import numpy as np
import matplotlib.pyplot as plt

def f1(g0,g1,t):
    r = 1e-9; B = 1e21
    return (g1/(g0+g1))*(1-e**(-r*B*((g0+g1)/g0)*t))

X = np.arange(0,30e-13,1e-16)
Y1 = f1(1,1,X)
Y2 = f1(1,3,X)
Y3 = f1(3,1,X)

plt.plot(X,Y1)
plt.plot(X,Y2)
plt.plot(X,Y3)

plt.legend(["g0 = 1, g1 = 1", "g0 = 1, g1 = 3", "g0 = 3, g1 = 1 (unlikely case)"])
plt.title("Population Fraction v/s Time \n (Typical values: rho = 1e-9, B = 1e21)")

plt.show()


```
